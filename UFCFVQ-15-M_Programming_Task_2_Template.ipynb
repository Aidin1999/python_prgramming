```python
# Select relevant columns from the 'result' DataFrame
test = result[['Entity', 'Bipolar disorder (%)', 'Eating disorders (%)']]

# Initialize an empty list to store correlation values
correlations_list = []

# Loop through each country and calculate correlation values
for country in test['Entity'].unique():
    
    # Subset the DataFrame for the specific country
    subset = test[test['Entity'] == country]
    
    # Calculate the correlation between 'Bipolar disorder (%)' and 'Eating disorders (%)'
    correlation_value = subset['Bipolar disorder (%)'].corr(subset['Eating disorders (%)'])
    
    # Append correlation values to the list
    correlations_list.append(correlation_value)

# Create a DataFrame from the list of correlation values
correlations = pd.DataFrame({
    'countries': test['Entity'].unique(),
    'cor': correlations_list
})

# Convert the 'cor' column to float type
correlations['cor'] = correlations['cor'].astype(float)

# Display the final DataFrame with country-wise correlation values
print(correlations)
```

```python
# 'correlations_list' is the list of correlation values, calculated above

# Create a figure with two subplots side by side
fig, axes = plt.subplots(1, 2, figsize=(12,8))

# Vertical Boxplot
sns.boxplot(y=correlations_list, orient='v', ax=axes[0])
axes[0].set_title('Boxplot of Correlation Values')
axes[0].set_ylabel('Correlation Values')
axes[0].set_xlabel('countries')


# Categorical Scatter Plot (Point Plot)
sns.scatterplot(x=range(len(correlations_list)), y=correlations_list, ax=axes[1])
axes[1].set_title('Scatter Plot of Correlations')
axes[1].set_xlabel('countries')
axes[1].set_xticks([])  # Remove x-axis ticks since the number for name of countries is point less



# Display the plots
plt.show()
```

### Requirement FR8 - Detect and remove any outliers in the data used for your "interesting" pattern or trend

## Outlier Detection using Z-Score

Detecting outliers is crucial for accurate analysis, and one effective method is using the z-score. Here are some key points:

1. **Symmetrical Data Range:**
   - The z-score is well-suited for data with a symmetrical range in both negative and positive zones (e.g., -5 to 5 or -6.2 to 6.2).

2. **Mean and Standard Deviation:**
   - It identifies outliers as data points far from the mean and exhibiting high standard deviation. Inliers, in contrast, show less difference from the overall mean.

3. **Application to Correlation Analysis:**
   - The z-score is particularly useful for correlation analysis, such as examining the correlation between eating and bipolar disorders across different countries.

4. **Inlier Detection:**
   - By using the z-score, we can identify inliers that closely align with the overall trend. This allows for a more generalized statement about the correlation.

5. **Impact on Trends:**
   - Some countries with different trends might be dropped, but the remaining countries tend to follow a similar trend. This strengthens the observed correlation between eating and bipolar disorders in the inlier countries.

6. **Z-Score Threshold for Inliers:**
   - The z-score threshold ensures that inliers have minimal deviation from the mean. This implies that the maximum and minimum values of inliers are as close as possible.

7. **Optimized Statement:**
   - Utilizing the z-score provides an optimized and inlier-focused approach for making statements about the correlation between eating and bipolar disorders across countries.


```python
import numpy as np
from scipy.stats import zscore

# 'column_to_check' is the column containing data between -1 and 1
column_to_check = 'cor'

# Create a copy of the DataFrame
correlations_optimized = correlations.copy()

# Calculate Z-scores
correlations_optimized['z_score'] = np.abs(zscore(correlations_optimized[column_to_check])) #it is cor 

z_score_threshold = 1 #it can be tunned but with one it showes more similar trends

# Identify outliers based on the Z-scores
outliers = correlations_optimized[correlations_optimized['z_score'] > z_score_threshold]

# Display the outliers
print("Outliers:")
print(outliers)

# Remove outliers from the DataFrame
correlations_optimized = correlations_optimized[correlations_optimized['z_score'] <= z_score_threshold]

# Display the DataFrame without outliers
print("\nDataFrame without outliers(inliners):")
print(correlations_optimized)
```

### Requirement FR9 - Define a hypothesis to test your â€œinterestingâ€ pattern or trend

## Hypothesis

**Hypothesis:**

- **Null Hypothesis:** I expect there are strong correlations between Eating disorders and bipolar disorder, excluding 43 countries (18.6% of all).
- **Alternative Hypothesis:** Globally, there might not be an overall correlation; it may vary case by case.

**Objective:**

- **Overall Goal:** To challenge and reject the Null Hypothesis in favor of the Alternative Hypothesis.
- If rejected, it suggests some countries have high correlations (more than 10% of all). In this case, we need to investigate each country, as some may heavily influence global records, potentially leading to misleading information.

**Testing Approach:**

- **Method:** One-sided t-sample test to determine if the correlation is greater.


## Precautions:

- **Caution with Null Hypothesis:** Even if p-value is close to 1, we can't fully accept the Null Hypothesis. Failing to reject it doesn't confirm its truth; we simply lack enough evidence.
- **P-Value Interpretation:** A high p-value and failure to reject the null hypothesis suggest a high certainty (e.g., 99.99%) that it's not disproven. But it doesn't confirm the null hypothesis.
- **Intent to Prove Null Hypothesis:** Stating "intent to prove" means we lack evidence to reject it, not that it's absolutely correct.
- **Continuous Investigation:** Ongoing process; results should be interpreted cautiously. Continuous investigation and considering other factors are crucial for robust conclusions.


### Requirement FR10 - Test your hypothesis with statistical significance level of 0.05

## Hypotheses for Correlation

**Null Hypothesis:**
- There is a correlation greater than 0.4 in these 188 countries.

**Alternative Hypothesis:**
- The correlation is not greater than 0.4 in these 188 countries.


```python
from scipy.stats import ttest_1samp#right library for this  test 

correlation_coefficient = correlations_optimized['cor']  # this the corr of 188 countries

# Transform the correlation coefficient to the z scale using arctanh (Fisher transformation)
z_transformed = np.arctanh(correlation_coefficient)#for having better result scaling and transfroming is great job

# Perform a one-sample t-test against a null hypothesis value greater than 0.4
null_hypothesis_value = 0.4
t_statistic, p_value = ttest_1samp(z_transformed, null_hypothesis_value,alternative='greater')

# Define the significance level (0.05)
alpha = 0.05

# Check if the p-value is less than the significance level
if p_value < alpha:
    print(f"Reject the null hypothesis. The correlation is not greater than {null_hypothesis_value}. p-value: {p_value}")
    print('Alternative Hypothesis true: Globally, there might not be an overall correlation; it may vary case by case.')
else:
    print("Fail to reject the null hypothesis. There is no significant difference.")
```

- It is essential to conduct a detailed study for each country on a case-by-case basis to understand the correlation more accurately.
